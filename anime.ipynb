```python
tmp_genre = []
for i in genre:
    if type(i) == list:
        for j in i:
            j = j.strip().replace(' ','_').replace(chr(10),'_').replace('-','_')
            if j not in tmp_genre:
                tmp_genre.append(j)
genre = tmp_genre
```

```python
len(tmp_genre)
```

```python
print(f"El total de generos de anime es {len(genre)}")
```

```python
# cuento la cantidad de anime por categoría y sumo los ratings de esas peliculas
for j in genre:
    print(j)
    anime[j+'_gen'] = np.where(j in anime['genre_list'], 1, 0)
```
